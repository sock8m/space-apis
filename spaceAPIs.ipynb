```python
import requests
```

```python
response = requests.get('https://ll.thespacedevs.com/2.2.0/launch/?format=json')
print(response)
print(response.json())
```

```python
response = requests.get('https://ll.thespacedevs.com/2.2.0/agencies/?format=json')
print(response)
print(response.json())
```

```python
URL = 'https://ll.thespacedevs.com/2.2.0/launch/'
PARAMS = {'mode':'list','search':'SpaceX'}
response = requests.get(url = URL, params = PARAMS)
print(response)
print(response.json())
```
