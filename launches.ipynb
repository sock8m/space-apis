```python
import requests
```

```python
URL = 'https://ll.thespacedevs.com/2.2.0/launch/upcoming'
PARAMS = {'mode':'list','limit':'5'}
response = requests.get(url = URL, params = PARAMS)
data = response.json()
```

```python
data
```

```python
import pandas as pd
launches = pd.DataFrame(data['results'])
launches
```

```python
launches = launches[['window_start','location','name']]
launches
```
